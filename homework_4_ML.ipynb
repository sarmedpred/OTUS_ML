```python
# Если вектора линейно независимы, то они могут быть базисными векторами. 
# Для этого создадим матрицу из векторов f1 и f2, и найдем определитель нового базиса.
C = np.array([[2,-1], 
             [-5,3]])

detC = np.linalg.det(C)
print(detC) # Определитель матрицы отличен от нуля, значит вектора могут быть базисными.

#B - канонический базис 
B = np.array([[1,0], 
             [0,1]])
x = np.array([[1], 
             [-4]])

# Найдем координаты вектора х в новом базисе, предварительно найдем обратную матрицу для С
Сinv = np.linalg.inv(C)
D = Сinv.dot(B)
xcoord = D.dot(x)
print(xcoord)
```
